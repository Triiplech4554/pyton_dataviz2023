#### ข้อ 4 Core Budget ไหน ('12-เงินสำรองจ่ายทั่วไป กรณีค่าใช้จ่ายต่างๆ เกี่ยวกับน้ำท่วม',
 '17-เงินสำรองจ่ายทั่วไป กรณีค่าใช้จ่ายเพื่อการพัฒนากรุงเทพมหานคร',
 '28-เงินสำรองค่าใช้จ่าย ในการพัฒนาเขต')  มีค่าเฉลี่ยงบประมาณที่ถุกอนุมัติสูงที่สุด

````python
CB_data1 = data1.dropna()
CoreBudget_data1 = CB_data1[data1['Core_budget_code'] != '-']
CoreBudget12 = data1[(data1['Core_budget_code'] == "12-เงินสำรองจ่ายทั่วไป กรณีค่าใช้จ่ายต่างๆ เกี่ยวกับน้ำท่วม")]
CoreBudget28 = data1[(data1['Core_budget_code'] == "28-เงินสำรองค่าใช้จ่าย ในการพัฒนาเขต")]
CoreBudget17 = data1[(data1['Core_budget_code'] == "17-เงินสำรองจ่ายทั่วไป กรณีค่าใช้จ่ายเพื่อการพัฒนากรุงเทพมหานคร")]
corename = []
````

````python
x=set(CoreBudget_data1['Core_budget_code'])
for n in x:
  corename.append(n)
CoreBudget12_MEAN =CoreBudget12['Approved_buget'].mean()
CoreBudget28_MEAN =CoreBudget28['Approved_buget'].mean()
CoreBudget17_MEAN =CoreBudget17 ['Approved_buget'].mean()
````

````python
if (CoreBudget12_MEAN > CoreBudget28_MEAN) & (CoreBudget12_MEAN > CoreBudget17_MEAN ):
  print(f'รายการนี้ {corename[0]} มีค่าเฉลี่ยงบประมาณที่ถูกอนุมัติสูงที่สุด {CoreBudget12_MEAN}')
elif (CoreBudget28_MEAN> CoreBudget12_MEAN ) & (CoreBudget28_MEAN > CoreBudget17_MEAN ):
  print(f'รายการนี้ {corename[1]} มีค่าเฉลี่ยงบประมาณที่ถูกอนุมัติสูงที่สุด {CoreBudget28_MEAN}')
elif (CoreBudget17_MEAN > CoreBudget12_MEAN ) & (CoreBudget17_MEAN > CoreBudget28_MEAN ):
  print(f'รายการนี้ {corename[2]} มีค่าเฉลี่ยงบประมาณที่ถูกอนุมัติสูงที่สุด {CoreBudget17_MEAN}')
````
