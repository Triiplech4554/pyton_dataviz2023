1.   งบประมาณประจำหรืองบประมาณลงทุน สูงกว่ากัน

````python
Fb = sep_oct_df[(sep_oct_df['รายการ']=='ประจำ')&( sep_oct_df['เดือน-ปี']=='sep-65')]['งบประมาณ']
Ib = sep_oct_df[(sep_oct_df['รายการ']=='ลงทุน')&( sep_oct_df['เดือน-ปี']=='sep-65')]['งบประมาณ']
print(f'งบประมาณประจำ มีค่าเท่ากับ {int(Fb):,} บาท')
print(f'งบประมาณลงทุน มีค่าเท่ากับ {int(Ib):,} บาท')
print(f'งบประมาณประจำ มีค่าสูงกว่า งบประมาณลงทุนอยู่ {(int(Fb) - int(Ib)):,} บาท')
````

2.   งบรวม เดือน oct หรือ sep อันไหนมากกว่ากัน

````python
sep = sep_oct_df[sep_oct_df["เดือน-ปี"] == sep_df["เดือน-ปี"][2]]
oct = sep_oct_df[sep_oct_df["เดือน-ปี"] == oct_df["เดือน-ปี"][2]]
Fb_sep  = sep['งบประมาณ'][2]
Ib_sep  = sep['งบประมาณ'][3]
sep_sum = Fb_sep + Ib_sep
Fb_oct = oct['งบประมาณ'][2]
Ib_oct= oct['งบประมาณ'][3]
oct_sum = Fb_oct + Ib_oct
list_name =[]
````

````python
for x in set(oct['รายการ']):
  list_name.append(x)
list_name
if sep_sum > oct_sum :
  print (f'งบรวมมากที่สุดคือเดือน {sep["เดือน-ปี"][2]} ประกอบไปด้วยงบ "{list_name[0]}" เป็นเงินจำนวน {int(Fb_sep):,} บาท เเละงบ "{list_name[1]}" เป็นเงินจำนวน {int(Ib_sep):,} บาท รวมเป็นเงิน {int(sep_sum):,} บาท')
elif sep_sum < oct_sum :
  print (f'งบรวมมากที่สุดคือเดือน {oct["เดือน-ปี"][2]} ประกอบไปด้วยงบ "{list_name[0]}" เป็นเงินจำนวน {int(Fb_oct):,} บาท เเละงบ "{list_name[1]}" เป็นเงินจำนวน {int(Ib_oct):,} บาท รวมเป็นเงิน {int(oct_sum):,} บาท')
````
